# Autograd

```python
# Numpy
import numpy as np

# PyTorch
import torch
import torch.nn as nn
import torch.nn.functional as F

# Visualización de grafo ade autograd
# https://github.com/szagoruyko/pytorchviz
from torchviz import make_dot, make_dot_from_trace
```

## Ejemplo de operaciones diferenciables

```python
# Creo un tensor
x = torch.tensor([1., 1.], requires_grad=True)
x
```

```python
# Opero
y=x**2
z=x**3
y,z
```

```python
r=(y+z).sum()
r
```

```python
# Grafo de operaciones
make_dot(r)
```

```python
# Computamos el gradiente
r.backward()
x.grad
```

## Efecto de utilizar .detach()

```python
# Qué pasa si usamos detach?
x = torch.tensor([1., 1.], requires_grad=True)
x
```

```python
y=x.detach()**2
z=x**3
y,z
```

```python
r=(y+z).sum()
r
```

```python
# Grafo de operaciones
# La rama de la operación que utilizó detach no lleva un tracking de las operaciones
make_dot(r)
```

```python
# Computamos el gradiente
r.backward()
x.grad
```

## Y si operamos con funciones no diferenciables?

```python
x = torch.tensor([1., 5.], requires_grad=True)
x
```

```python
# Funcion no diferenciable: Max. 
y = x.max()
y
```

```python
make_dot(y)
```

```python
# Computamos el gradiente
y.backward()
y.grad

if y.grad is None:
    print('None')
```
